```python
# Importing all The libraries all the necessary ones 
from bs4 import BeautifulSoup
from collections import Counter
import urllib.request
import urllib.parse,urllib.error
import ssl
import requests
import csv
import sqlite3 as db
# Importing all the ctx that are certificates used to pass the security check.
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
#header: Describing yourself as one of the browser while fetching the value
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
AllMoviesName = []
linkOfMovies = []
#Loop all the pages to get data from first five pages
for i in range(5):
    # Defining the url and then appending the page number 
    url = 'https://www.metacritic.com/browse/movies/score/metascore/all/filtered?sort=desc&page=' + str(i)
    individualPage = requests.get(url, headers = headers)
    #Getting the html element from the page
    containerData = BeautifulSoup(individualPage.content, "html.parser")
    #going in clamp-summary-wrap to get the movies name from the element
    gettingTheContainer = containerData.find_all('td', class_ = 'clamp-summary-wrap')
    
    for movie in gettingTheContainer:
        #in h3 element get the movie name in text
        AllName = movie.find('h3').text
        #appending all the data in the list
        AllMoviesName.append(AllName)
        #In the a tag there is title of the movie and fetching it.
        titletagData = movie.find('a', class_ = 'title')
        AllTheLink = titletagData.get('href', None)
        # Getting all the final movies names
        linkOfMovies.append(AllTheLink)
print(AllMoviesName)

```

```python
# From the size of the movie_link all the movies name append it to metacritic.com to get the full page movie link
for i in range(len(linkOfMovies)):
    linkOfMovies[i] = 'https://www.metacritic.com'+ linkOfMovies[i]
print(linkOfMovies)
```

```python
#Creating a dictionary
AllTheDetailLink = []
#for loop to go through all the movies link
for mov in linkOfMovies:
    
    individualPage = requests.get(mov, headers = headers)
    #now in individual movie page go and fetch the data
    containerData = BeautifulSoup(individualPage.content, "html.parser")
    #There is container see_all pad_btm1 in this page which we can see using inspecting each element
    div = containerData('div', class_= 'see_all pad_btm1')
    #looping through each div
    for line in div:
        # in each div there is 'a' tag container that we need to seach to extract the data.
        TextHyperLink = line.find('a')
        #getting the URL of detail page 
        url = TextHyperLink.get('href', None)
        AllTheDetailLink.append(url)
```

```python
for i in range(len(AllTheDetailLink)):
    #appending the each movie with the detail link at the end.
    AllTheDetailLink[i] = 'https://www.metacritic.com'+ AllTheDetailLink[i]
print(AllTheDetailLink)
```

```python
movie_director = dict()
total_castValues = []
total_directorValues = []
#iterating through the length of the detail page
for i in range(len(AllTheDetailLink)):
    page = requests.get(AllTheDetailLink[i], headers = headers)
    #scraping the data through the main html container
    soup = BeautifulSoup(page.content, "html.parser")
    add = '\'s'
    #If movies name ends with s add them as well. We faced some challenge here since the movies that ends with s were not being appended
    #creating that a problem and making the total list of the movies count very less
    if AllMoviesName[i].endswith('s'):
        add = '\''
    #extracting the director names of each movie and putting it in below table.
    DirectorDataTable = soup.find('table', class_ = 'credits', summary = AllMoviesName[i] +add+" Director Credits")
        #extracting the cast names of each movie and putting it in below table.
    AllCastData = soup.find('table', class_ = 'credits', summary = AllMoviesName[i] +add+" Cast Credits")
        #extracting the p_cast names of each movie and putting it in below table.
    PrincipalCastAllData = soup.find('table', class_ = 'credits', summary = AllMoviesName[i] + "'s Principal Cast Credits")

    directors = list()
    casts = list()
    
    #If director table is not then getting the none value which use to give error as none type.
    #So for that solution i appended the 'director not available' in the list and continuing the operations
    if DirectorDataTable is None:
        total_directorValues.append('director not available')
        directors.append('director not available')
    #if DirectorDataTable is not none then append it to the directors and also total_directorValues 
    else:
        director_line = DirectorDataTable.find_all('td', class_ = 'person')

        for line in director_line:
            TextHyperLink = line.find('a')
            director = TextHyperLink.get_text()
            directors.append(director.strip())
            total_directorValues.append(director.strip())
        
    #If cast table is not then getting the none value which use to give error as none type.
    #So for that solution i appended the 'director not available' in the list and continuing the operations
    if AllCastData is None:
        casts.append('cast not available')
    elif AllCastData is None:
        cast_line = AllCastData.find_all('td', class_ = 'person')
        for line in cast_line:
            TextHyperLink = line.find('a')
            cast = TextHyperLink.get_text()
            casts.append(cast.strip())
            total_castValues.append(cast.strip())
        #if PrincipalCastAllData is not none then append it to the casts and also total_castValues 
    elif PrincipalCastAllData is not None:
        pLine = PrincipalCastAllData.find_all('td', class_ = 'person')
        for line in pLine:
            TextHyperLink = line.find('a')
            past = TextHyperLink.get_text()
            casts.append(past.strip())
            total_castValues.append(past.strip())
            
            
    
    
    director_cast = dict()
    diffVal = set(casts)
    diffList = list(diffVal)
    
    director_cast[tuple(directors)] = diffList
    #Creating a movie_director dictionary and make its key as director and values as cast they have worked with
    movie_director[AllMoviesName[i]] = director_cast

print(movie_director)




```
